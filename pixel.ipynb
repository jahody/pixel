## Image

```python
img = Image.open("ren2.jpg")
```

```python
img.size
```

## Resize

```python
factor = 50
```

```python
width = img.size[0]//factor
```

```python
height = img.size[1]//factor
```

```python
width
```

```python
height
```

```python
imgSmall = img.resize((width,height),resample=Image.BILINEAR)
```

```python
imgSmall
```

```python
pix_val = list(imgSmall.getdata())
```

```python
len(pix_val)
```

## Greyscale

```python
imgBW = imgSmall.convert("L")
```

```python
imgBW
```

```python
pix_valBW = list(imgBW.getdata())
```

```python
len(pix_valBW)
```

```python
mx = max(pix_valBW)
mx
```

```python
mn = min(pix_valBW)
mn
```

```python
MX = mx - mn
MX
```

```python
def Average(lst): 
    return sum(lst) // len(lst) 
```

```python
Average(pix_valBW)
```

## Colors

```python
colors = 8

col8 = (233,234,160)
col7 = (180,233,231)
col6 = (177,218,233)
col5 = (158,206,204)
col4 = (73,75,35)
col3 = (69,68,63)
col2 = (31,60,68)
col1 = (30,44,71)
```

## Asign

```python
imgNew = []
```

```python
len(pix_valBW)
```

```python
for e in pix_valBW:
    if e <= MX//colors:
            imgNew.append(col1)
            
    if e > MX//colors and e <= MX//colors*2:
            imgNew.append(col2)
            
    if e > MX//colors*2 and e <= MX//colors*3:
            imgNew.append(col3)
            
    if e > MX//colors*3 and e <= MX//colors*4:
            imgNew.append(col4)
            
    if e > MX//colors*4 and e <= MX//colors*5:
            imgNew.append(col5)
            
    if e > MX//colors*5 and e <= MX//colors*6:
            imgNew.append(col6)
            
    if e > MX//colors*6 and e <= MX//colors*7:
            imgNew.append(col7)
            
    if e > MX//colors*7:
            imgNew.append(col8)
```

```python
len(imgNew)
```

## New image

```python
im2 = Image.new(imgSmall.mode, imgSmall.size)
```

```python
im2.putdata(imgNew)
```

```python
im2.save("render_new.png")
```
